```python
# read in the data
import pandas as pd
url = 'titanic.csv'
titanic = pd.read_csv(url)
```

```python
# fill in the missing values for age with the median age
titanic.Age.fillna(titanic.Age.median(), inplace=True)
titanic.Embarked.fillna('S', inplace=True)
```

```python
# encode female as 0 and male as 1
titanic.Sex = titanic.Sex.map({'female':0, 'male':1})
```

```python
# create a DataFrame of dummy variables for Embarked
embarked_dummies = pd.get_dummies(titanic.Embarked, prefix='Embarked', drop_first=True)
titanic
```

```python
# concatenate the original DataFrame and the dummy DataFrame
titanic = pd.concat([titanic, embarked_dummies], axis=1)

# print the updated DataFrame
titanic.head()
```

Survived: 0=died, 1=survived (response variable)

Pclass: 1=first class, 2=second class, 3=third class

Sex: 0=female, 1=male

Age: numeric value

Embarked: C or Q or S

```python
# define X and y
feature_cols = ['Pclass', 'Sex', 'Age', 'Embarked_Q', 'Embarked_S']
X = titanic[feature_cols]
y = titanic.Survived
```

```python
# fit a classification tree with max_depth=3 on all data
from sklearn.tree import DecisionTreeClassifier
treeclf = DecisionTreeClassifier(max_depth=3, random_state=1)
treeclf.fit(X, y)
```

### 决策树的可视化

可视化决策树需要先安装python-graphviz库，使用conda安装的命令如下：
conda install python-graphviz

然后再把..\Anaconda3\Library\bin\graphviz目录添加到Windows系统的环境变量Path中

```python
import graphviz
from sklearn import tree

dot_data = tree.export_graphviz(treeclf, out_file=None, 
                         feature_names=feature_cols,
                         class_names="Survived",  
                         filled=True, rounded=True,  
                         special_characters=True)
graph = graphviz.Source(dot_data)
# graph.render("titanic")
graph
```

非叶节点：

samples: 分裂前该节点包含的样例个数

gini: 该节点的基尼指数

rule: 用于分裂该节点的规则（为真走左子树，为假走右子树）

value: 该节点包含的负例的个数和正例的个数

class: 该节点预测的类别

叶子节点：

samples: 该节点包含的样例个数

value: 该节点包含的负例的个数和正例的个数

gini: 该节点的基尼指数

class: 该节点预测的类别


注意最右下的分裂，两个子节点预测的是相同的类。该分类没有改变分类错误率，但是它提高了类别的纯度，这样可以提高我们分类预测的概率。
